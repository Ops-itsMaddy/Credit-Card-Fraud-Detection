**Importing the Dependencies**

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#Importing the dataset into Panda's Dataframe
credit_card_dataset = pd.read_csv('/content/creditcard.csv')
```

```python
#Retrieve the first five rows of Dataset
credit_card_dataset.head()
```

```python
#Retrieve the lat five rows od Dataset
credit_card_dataset.tail()
```

```python
#Dataset Information
credit_card_dataset.info()
```

```python
#Checking if any columns has missing values
credit_card_dataset.isnull().sum()
```

```python
#Distribution of Legit and Fraudulent Transactions
credit_card_dataset['Class'].value_counts()
```

This Dataset is highly Unbalanced.

0 --> Normal Transaction

1 --> Fraudulent Transaction

```python
#Separating the Data for Analysis
legit = credit_card_dataset[credit_card_dataset.Class == 0]
fraud = credit_card_dataset[credit_card_dataset.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#Statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#Compare the values of both transactions
credit_card_dataset.groupby('Class').mean()
```

**Under-Sampling**

Build a Sample Dataset containing similar distribution of Legit Transactions and Fraudulent Transactions.

Number of Fraudulent Transactions --> 492

```python
legit_sample = legit.sample(n = 492)
```

**Concatenating Two DataFrames**

```python
new_dataset = pd.concat([legit_sample, fraud], axis = 0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the Data into Features and Labels

```python
X = new_dataset.drop(columns = 'Class', axis = 1)
Y = new_dataset['Class']
```

```python
print(X)
print(Y)
```

Spliting the data into Training and Testing Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.2, random_state= 2, stratify= Y)
```

```python
print(X.shape ,X_train.shape, X_test.shape)
```

**Model Training**

Logistic Regression

```python
model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

**Model Evaluation**

Accuracy Score

```python
#Accuracy on Training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print("Accuracy of the Training Data :",training_data_accuracy * 100 )
```

```python
#Accuracy on the Testing Data
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print("Accuracy of the Testing Data :", testing_data_accuracy * 100 )
```
